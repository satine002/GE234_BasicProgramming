## 🔹 ตัวอย่างที่ 1: ใช้ NumPy คำนวณข้อมูล NDVI จากภาพ Raster


```python

import numpy as np

# สร้างข้อมูลตัวอย่างสำหรับภาพ NDVI (Normalized Difference Vegetation Index)
nir = np.array([[0.7, 0.8, 0.6], [0.9, 0.5, 0.3], [0.4, 0.7, 0.2]])
red = np.array([[0.3, 0.4, 0.2], [0.5, 0.3, 0.1], [0.2, 0.3, 0.1]])

# คำนวณค่า NDVI
ndvi = (nir - red) / (nir + red)

print("ค่า NDVI:")
print(ndvi)
```


## 🔹 ตัวอย่างที่ 2: ใช้ NumPy คำนวณค่าเฉลี่ยและค่ามากสุดของ NDVI


```python

print(f"ค่าเฉลี่ย NDVI: {np.mean(ndvi):.2f}")
print(f"ค่า NDVI สูงสุด: {np.max(ndvi):.2f}")
print(f"ค่า NDVI ต่ำสุด: {np.min(ndvi):.2f}")
```


## 🔹 ตัวอย่างที่ 3: ใช้ GeoPandas โหลดและวิเคราะห์ข้อมูล Shapefile


```python

import geopandas as gpd

# โหลดข้อมูลชั้นข้อมูลจังหวัดของประเทศไทย
gdf = gpd.read_file("https://data.opendevelopmentmekong.net/dataset/64ee3517-4e16-418f-81da-722c123ddacd/resource/ed6ff6ec-fbe6-43a5-99f1-e057c2796c33/download/th_province_boundary.shp")

# แสดงข้อมูล 5 แถวแรก
print(gdf.head())
```


## 🔹 ตัวอย่างที่ 4: คำนวณพื้นที่ของแต่ละจังหวัด


```python

# ตรวจสอบค่า CRS (Coordinate Reference System)
print(gdf.crs)

# คำนวณพื้นที่ของแต่ละจังหวัด (หน่วยเป็นตารางเมตร)
gdf["area_sqkm"] = gdf.geometry.area / 1e6

# แสดงพื้นที่จังหวัด 5 อันดับแรกที่ใหญ่ที่สุด
print(gdf.nlargest(5, "area_sqkm")[["PROV_NAME", "area_sqkm"]])
```


## 🔹 ตัวอย่างที่ 5: ใช้ GeoPandas ทำ Spatial Join


```python

# โหลดข้อมูลชั้นข้อมูลอำเภอ
gdf_districts = gpd.read_file("https://data.opendevelopmentmekong.net/dataset/ed8e818c-e1e3-4746-a7cd-6e437c078b9d/resource/9db5a579-2673-4e83-bf4f-bc7a83316c35/download/th_district_boundary.shp")

# ทำ Spatial Join ระหว่างอำเภอกับจังหวัด
gdf_joined = gpd.sjoin(gdf_districts, gdf, how="inner", predicate="within")

# แสดงตัวอย่างข้อมูลที่เชื่อมโยงกัน
print(gdf_joined.head())
```

**แบบฝึกหัด 1**: ใช้ NumPy คำนวณค่า **Mean, Max, Min** ของค่า NDVI ในอาร์เรย์ที่สร้างขึ้นเอง

```python
import numpy as np

# สร้างข้อมูล NDVI ตัวอย่างขึ้นมาเอง (สามารถปรับเปลี่ยนค่าได้)
# ใช้ค่า NDVI จากตัวอย่างที่ 1 หรือสร้างใหม่ได้
ndvi_custom = np.array([
    [0.4, 0.33, 0.5],
    [0.28, 0.25, 0.5],
    [0.33, 0.4, 0.33]
])

print(f"อาร์เรย์ NDVI ที่สร้างขึ้นเอง:\n{ndvi_custom}")
print(f"\nค่าเฉลี่ย NDVI: {np.mean(ndvi_custom):.2f}")
print(f"ค่า NDVI สูงสุด: {np.max(ndvi_custom):.2f}")
print(f"ค่า NDVI ต่ำสุด: {np.min(ndvi_custom):.2f}")
```

**แบบฝึกหัด 2**: ใช้ GeoPandas โหลด **Shapefile** ของจังหวัด และคำนวณพื้นที่ของแต่ละจังหวัด

```python

# ตรวจสอบค่า CRS (Coordinate Reference System) - แสดงผล CRS ของ gdf
print(f"CRS ของข้อมูลจังหวัด: {gdf.crs}")

# คำนวณพื้นที่ของแต่ละจังหวัด (หน่วยเป็นตารางเมตร)
# GeoPandas จะคำนวณพื้นที่โดยอัตโนมัติเป็นหน่วยของ CRS นั้นๆ
# ตรวจสอบว่า CRS เป็น Projected CRS หรือไม่
# ถ้าเป็น Geographic CRS (เช่น EPSG:4326) ควรแปลงเป็น Projected CRS ก่อน

gdf["area_sqkm"] = gdf.geometry.area / 1e6 # แปลงหน่วยเป็นตารางกิโลเมตร

# แสดงพื้นที่ 5 จังหวัด
print("\nพื้นที่ 5 จังหวัด:")
print(gdf.nlargest(5, "area_sqkm")[["PROV_NAMT", "area_sqkm"]])
```

**แบบฝึกหัด 3**: ใช้ GeoPandas ทำ **Spatial Join** ระหว่างข้อมูลจังหวัดและอำเภอ

```python
import geopandas as gpd

# โหลดข้อมูลชั้นข้อมูลอำเภอจาก Google Drive
gdf_districts = gpd.read_file("/content/drive/MyDrive/Amphoe_PROV", encoding='TIS-620')

# ตรวจสอบและแปลง CRS ของ gdf_districts ให้ตรงกับ gdf
# ถ้า gdf_districts.crs ไม่ตรงกับ gdf.crs ให้แปลง
if gdf_districts.crs != gdf.crs:
    gdf_districts = gdf_districts.to_crs(gdf.crs)

# ทำ Spatial Join ระหว่างอำเภอกับจังหวัด
gdf_joined = gpd.sjoin(gdf_districts, gdf, how="inner", predicate="within")

# แสดงตัวอย่างข้อมูลอำเภอและจังหวัดที่ถูก Join กันแล้ว
print(gdf_joined[['AMP_NAME_T', 'PROV_NAMT']].head())
```

**แบบฝึกหัด 4**: ใช้ NumPy และ GeoPandas ร่วมกันเพื่อหาข้อมูลจังหวัดที่มี NDVI เฉลี่ยสูงสุด


```python
import numpy as np

# ตรวจสอบว่า gdf มีข้อมูลอยู่หรือไม่
if 'gdf' not in locals():
    print("ไม่พบ GeoDataFrame 'gdf' กรุณารันเซลล์ที่โหลดข้อมูลจังหวัดก่อน")
else:
    # เนื่องจากไม่มีข้อมูล จึงใช้วิธีการสุ่ม สมมติค่า NDVI เฉลี่ยของจังหวัดขึ้นมาเอง
    np.random.seed()
    gdf['avg_ndvi'] = np.random.uniform(0.1, 0.9, len(gdf))

    # หาจังหวัดที่มีค่า NDVI เฉลี่ยสูงสุด
    province_max_ndvi = gdf.loc[gdf['avg_ndvi'].idxmax()]

    print("\nจังหวัดที่มีค่า NDVI เฉลี่ยสูงสุด:")
    print(f"ชื่อจังหวัด (ภาษาไทย): {province_max_ndvi['PROV_NAMT']}")
    print(f"ชื่อจังหวัด (ภาษาอังกฤษ): {province_max_ndvi['PROV_NAME']}")
    print(f"ค่า NDVI เฉลี่ย: {province_max_ndvi['avg_ndvi']:.2f}")
```
